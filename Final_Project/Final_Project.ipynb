```python
##MP3 Veri Toplama
import json
import pandas as pd

##Veri Yükleme
##Spotipy kütüphanesi kullanılarak toplanan oturum verisinin JSON dosya yolu.
##Lütfen bu yolu kendi dosya yapınıza göre güncelleyin.
path = "data\sessions\session_20251203_184308_Canımsın.json"

try:
    with open(path, "r", encoding="utf-8") as f:
        session = json.load(f)

    tracks = session["tracks"]
    df = pd.json_normalize(tracks)
    print("Veri başarıyla yüklendi ve DataFrame'e dönüştürüldü.")

except FileNotFoundError:
    print(f"HATA: Veri dosyası '{path}' bulunamadı. Lütfen dosya yolunu kontrol edin.")
    df = pd.DataFrame()

##Temel Veri Ön İşleme
if not df.empty:
    # Sanatçı listesini tek bir dizeye dönüştürme (görüntüleme kolaylığı için)
    df["artists_str"] = df["artists"].apply(lambda xs: ", ".join(xs))

##Çıkış tarihinden yıl bilgisini çıkarma
    df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
    df["year"] = df["release_date"].dt.year

    print("\n--- İşlenmiş Verinin İlk 5 Satırı ---")
    display(df[["order_index", "name", "artists_str", "popularity", "year", "artist_genres_main"]].head())
```
