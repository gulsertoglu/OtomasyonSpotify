```python
import json
import pandas as pd

path = "data/sessions/session_20251203_184308_Canımsın.json"


with open(path, "r", encoding="utf-8") as f:
    session = json.load(f)

tracks = session["tracks"]

df = pd.json_normalize(tracks)
df.head(50)
```

```python
df[["order_index", "name", "artists", "popularity", "artist_genres_main"]]
```

```python
df["artists_str"] = df["artists"].apply(lambda xs: ", ".join(xs))
df[["order_index", "name", "artists_str", "popularity"]]
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
df["year"] = df["release_date"].dt.year
df[["name", "year", "popularity"]]
```

```python
from collections import Counter

all_genres = []
for g_list in df["artist_genres_main"]:
    if isinstance(g_list, list):
        all_genres.extend(g_list)

Counter(all_genres)
```
